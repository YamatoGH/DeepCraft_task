以下の理由から、ベースモデルの線形モデルとして"SARIMAX"を使用

-> 一定の(変数による)説明性を確保

-> 時系列を利用

-> データの確認時、季節性を確認

```python
#使用ライブラリを読み込み

import pandas as pd 
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import itertools
import warnings
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./use_data/ett.csv",parse_dates=['date'], index_col='date')
```

```python
df
```

データの分割は、

訓練データ -> 2018-01-31まで (データの8割)

テストデータ -> 2018-02-01以降 (データの2割)

で行う。検証データはこちらでは用いない。 (過適合とのトレードオフを考えると、線形モデルのため)

```python
# データの分割
train_data = df[:'2018-01-31 23:00:00']
train_data = train_data.asfreq('h')  # 時間単位のデータ

test_data = df['2018-02-01 00:00:00':]
```

```python
# 仮説を立てたパラメータでSARIMAXモデルを作る（周期を3で指定してやってみる）

exog_list = ['HUFL','HULL','MUFL','MULL','LUFL']  #外部変数リスト
model = SARIMAX(train_data['OT'],
                exog=train_data[exog_list],
                order=[2, 0, 1],
                seasonal_order=[2, 1, 1, 3],    #周期を3で指定
                enforce_stationarity=False,
                enforce_invertibility=False)

# モデルを適合
model_fit = model.fit(disp=False)

# 結果の表示
print(model_fit.summary())
```

```python
# 予測の実行と精度の評価(RMSEを使用)

pred = model_fit.predict(start=len(train_data), end=len(train_data)+len(test_data)-1,
                             exog=test_data[exog_list])

mse = mean_squared_error(test_data['OT'], pred)
rmse = np.sqrt(mse)
print('RMSE:', rmse)
```

```python
# 予測値と実測値をプロット

plt.figure(figsize=(12, 6))
plt.plot(test_data.index, test_data['OT'], label='Actual')
plt.plot(test_data.index, pred, label='Predicted', linestyle='--')
plt.legend()
plt.title('Actual vs Predicted')
plt.show()
```
